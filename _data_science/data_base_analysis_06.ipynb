## Criando novas variáveis

#### Importando bibliotecas

```python
import pandas as pd
```

#### Importando dados

```python
data = pd.read_csv('../data/aluguel_residencial.csv', sep=';')
data.head(10)
```

#### Nova variável - valor bruto

```python
data['Valor Bruto'] = data['Valor'] + data['Condominio'] + data['IPTU']
data.head(10)
```

#### Nova variável - valor m2

```python
data['Valor M2'] = (data['Valor'] / data['Area']).round(2)
data.head(10)
```

#### Nova variável - valor bruto m2

```python
data['Valor Bruto M2'] = (data['Valor Bruto'] / data['Area'] ).round(2)
data.head(10)
```

#### Criar uma nova variável com as variações de casa como casa, por exemplo: Casa, Casa de Condomínio e Casa de Vila serão consideradas casas

```python
casa = ['Casa', 'Casa de Condomínio', 'Casa de Vila']
```

```python
data['Tipo Agregado'] = data['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')
data.head(10)
```

#### Excluindo variáveis

* Primeira forma - DEL

```python
data_aux = pd.DataFrame(data[['Tipo Agregado', 'Valor Bruto', 'Valor Bruto M2']])
data_aux.head(10)
```

```python
del data_aux['Valor Bruto']
data_aux.head(10)
```

* Segunda forma - POP

```python
data_aux.pop('Valor Bruto M2')
data_aux.head()
```

* Terceira forma - DROP

```python
data.drop(['Valor Bruto', 'Valor Bruto M2'], axis=1, inplace=True)
data.head()
```

#### Exportando e sobre escrevendo arquivo anterior

```python
data.to_csv('../data/aluguel_residencial.csv', sep=';', index=False)
```
