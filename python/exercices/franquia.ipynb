### Problema

Criar um modelo de regressão linear para prever qual será o investimento inicial necessário de um franquia a partir da taxa anual cobrada pelo franquador

#### Importando bibliotecas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

#### Importando dados

```python
data = pd.read_csv('../../data/slr12.csv', sep=';')
data.shape
```

```python
data.head()
```

### Definição das variáveis X, Y

```python
x = data.iloc[:,1].values
y = data.iloc[:,0].values
```

#### Cálculo da correlação entre X e Y

```python
correlacao = np.corrcoef(x,y)
correlacao
```

#### Formatando para matriz

```python
x = x.reshape(-1, 1)
modelo = LinearRegression()
modelo.fit(x,y)
```

#### Visualização dos coeficientes

```python
modelo.intercept_
```

```python
modelo.coef_
```

#### Geração do gráfico com pontos reais e previsões

```python
plt.scatter(x,y)
plt.plot(x, modelo.predict(x), color='red')
plt.title('Relação entre investimento inicial e taxa anual')
plt.xlabel('Investimento inicial')
plt.ylabel('Taxa anual')
```
