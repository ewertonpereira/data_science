# Gráfico 3D
___

#### Importando bibliotecas

```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
import warnings
import numpy as np
warnings.filterwarnings('ignore')
```

#### Importando os dados

```python
data = pd.read_csv('../../data/orchard.csv')
data.head()
```

#### Criação do gráfico 3D, indicando o atributo projection = '3d' e passando três atributos (drecrease, rowpos, colpos)

```python
figure = plt.figure()
axis = figure.add_subplot(111, projection='3d') # or 1, 1, 1
axis.scatter(data.decrease, data.rowpos, data.colpos)
axis.set_xlabel('decrease')
axis.set_ylabel('rowpos')
axis.set_zlabel('colpos')
```
