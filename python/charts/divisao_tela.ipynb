## Divisão de tela (subgráficos)
___

#### Importando bibliotecas

```python
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

#### Importando dados 

```python
data = pd.read_csv('../data/trees.csv')
data.head()
```

#### Girth com volume

```python
plt.scatter(data.Girth, data.Volume)
```

#### Girth com weight

```python
plt.scatter(data.Girth, data.Height)
```

#### Height com volume

```python
plt.scatter(data.Height, data.Volume)
```

#### Histograma volume

```python
plt.hist(data.Volume)
```

#### Imprimindo juntos

Criação de figura, na qual os gráficos serão posicionados

```python
plt.figure(1)
plt.subplot(2, 2, 1)
plt.scatter(data.Girth, data.Volume)
plt.subplot(2, 2, 2)
plt.scatter(data.Girth, data.Height)
plt.subplot(2, 2, 3)
plt.scatter(data.Height, data.Volume, marker='.')
plt.subplot(2, 2, 4)
plt.hist(data.Volume)
```
