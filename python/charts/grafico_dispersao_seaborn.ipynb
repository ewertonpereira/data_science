# Gráfico de dirpersão com seaborn
___

#### Importando bibliotecas

```python
import pandas as pd
import seaborn as srn
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

#### Importando os dados

```python
data = pd.read_csv('../../data/co2.csv')
data.head()
```

#### Gráfico de dipersão utilizando os atributos conc e uptake, agrupando pelo type

```python
srn.scatterplot(x=data.conc, y=data.uptake, hue=data.Type)
```

#### Seleção de registros específicos da base de dados (Quebec e Mississipi) 

```python
q = data.loc[data['Type'] == 'Quebec']
m = data.loc[data['Type'] == 'Mississippi']
```

#### Subgráfico ( 1 linhas e 2 colunas) mostrando gráficos sobre cada região

```python
plt.figure()
plt.subplot(1, 2, 1)
srn.scatterplot(x=q.conc, y=q.uptake).set_title('Quebec')
plt.subplot(1, 2, 2)
srn.scatterplot(x=m.conc, y=m.uptake).set_title('Mississippi')
plt.tight_layout()
```

#### Refrigerado e não refrigerado

```python
ch = data.loc[data['Treatment'] == 'chilled']
nc = data.loc[data['Treatment'] == 'nonchilled']
```

#### Gráfico somente com 'chilled' e 'nonchilled'

```python
plt.figure()
plt.subplot(1, 2, 1)
srn.scatterplot(x=ch.conc, y=ch.uptake).set_title('Chilled')
plt.subplot(1, 2, 2)
srn.scatterplot(x=nc.conc, y=nc.uptake).set_title('Non chilled')
plt.tick_params()
```

#### Importando novo dado

```python
data2 = pd.read_csv('../../data/esoph.csv')
data2.head()
```

#### Gráficos entre atributos 'alcgp' e 'ncontrols'

```python
palette_colors = ['brown', 'orange', 'green', 'red']
srn.catplot(x='alcgp', y='ncontrols', data=data2, jitter=False, palette=srn.color_palette(palette_colors))
```

#### Gráficos entre atributos 'alcgp' e 'ncontrols', com agrupamento

```python
srn.catplot(x='alcgp', y='ncontrols', data=data2, col='tobgp', palette=srn.color_palette(palette_colors))
```
