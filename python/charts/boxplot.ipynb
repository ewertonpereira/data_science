## Boxplot
___

#### Importando bibliotecas

```python
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

#### Importando dados

```python
data = pd.read_csv('../data/trees.csv')
data.head()
```

#### Geração do boxplot

    patch_artist = True -> Preenche cor
    showfliers  = False -> Não mostra os outliers
    notch = True -> Faz o entalhe

```python
plt.boxplot(data.Volume, vert=False, showfliers=False, notch=True,  patch_artist=True)
plt.title('Árvores')
plt.xlabel('Volume')
```

#### Dados por linha

```python
plt.boxplot(data)
plt.title('Árvores')
plt.xlabel('Dados')
```

#### Geração de três boxplots, cada um mostrando informações diferentes

```python
plt.boxplot(data.Volume, vert=False)
plt.boxplot(data.Girth, vert=False)
plt.boxplot(data.Height, vert=False)
plt.title('Árvores')
plt.xlabel('Dados')
```
