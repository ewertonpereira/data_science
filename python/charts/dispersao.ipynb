## Gráfico de dispersão
___

#### Importando bibliotecas

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

#### Importando dados

```python
data = pd.read_csv('../data/trees.csv')
data.head()
```

#### Gráfico de dispersão considerando o volume e a dispersão

```python
plt.scatter(
    data.Girth,
    data.Volume,
    color='blue',
    facecolors='none',
    marker='.'
)
plt.title('Árvores')
plt.xlabel('Volume')
plt.ylabel('Circunferência')
```

#### Gráfico de linha considerando o volume e o atributo 'Girth'

```python
plt.plot(data.Girth, data.Volume)
plt.title('Árvores')
plt.xlabel('Volume')
plt.ylabel('Circunferência')
```

#### Gráfico de dispersão com 'afastamento' dos dados (jitter)
    fit_reg -> Linha de tendência
    jitter -> Não deixa dados sobrescrito

```python
sns.regplot(
    y='Girth',
    x='Volume',
    data=data,
    x_jitter=0.3,
    fit_reg=False)
```
