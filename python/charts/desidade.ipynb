## Gráficos de densidade
___

#### Importando bibliotecas

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

#### Importando dados

```python
data = pd.read_csv('../data/trees.csv')
data.head()
```

#### Histograma com divisões (bins) e somente para o primeiro atributo da base de dados

```python
plt.hist(data.iloc[:,1], bins=6)
```

#### Histograma com a linha de distribuição de frequência, com 6 divisões (bins)
kde -> linha de densidade

#### Histograma

```python
sns.distplot(
    data.iloc[:,1], 
    hist=True,
    kde=False,
    bins=6,
    color='blue',
    hist_kws={'edgecolor': 'black'}
)
```

#### Densidade

```python
sns.distplot(
    data.iloc[:,1], 
    hist=False,
    kde=True,
    bins=6,
    color='blue',
    hist_kws={'edgecolor': 'black'}
)
```

#### Densidade e histograma

```python
sns.distplot(
    data.iloc[:,1], 
    hist=True,
    kde=True,
    bins=6,
    color='blue',
    hist_kws={'edgecolor': 'black'}
)
```
