## Boxplot com Seaborn
___

#### Importando bibliotecas

```python
import pandas as pd
import seaborn as srn
import warnings
warnings.filterwarnings('ignore')
```

#### Importando dados

```python
data = pd.read_csv('../../data/trees.csv')
data.head()
```

#### Visualização de um boxplot

```python
srn.boxenplot(data.Volume).set_title('Árvores')
```

#### Visualização de vários boxplot na mesma imagem

```python
srn.boxenplot(data=data)
```
