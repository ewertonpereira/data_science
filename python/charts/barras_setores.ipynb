## Gráfico de barras e de setores
___

#### Importando bibliotecas

```python
import pandas as pd
```

#### Importando dados

```python
data = pd.read_csv('../../data/insect.csv')
data.head()
```

#### Agrupamento de dados baseados no atributo 'spray', contando e somando os registros

```python
grouping = data.groupby(['spray'])['count'].sum()
grouping
```

#### Gráfico de barras

```python
grouping.plot.bar(color='pink')
```

#### Cores

```python
grouping.plot.bar(color=['pink', 'blue', 'orange', 'green', 'red', 'yellow'])
```

#### Gráfico de pizza

```python
grouping.plot.pie()
```

#### Gráfico de pizza com legenda

```python
grouping.plot.pie(legend=True)
```
