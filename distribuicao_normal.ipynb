### Distribuição normal

#### Importando bibliotecas

```python
from scipy.stats import norm
import matplotlib.pyplot as plt
import numpy as np
```

### Conjunto de objetos em uma cesta, a média é 8 e o desvio padrão é 2.

#### Qual a probabilidade de tirar um objeto de peso menor que 6 quilos?

```python
norm.cdf(6, 8, 2)
```

```python
# Parâmetros da distribuição normal
mean = 8
std_dev = 2

# Ponto específico
x_point = 6
sf_value_at_x_point = norm.sf(x_point, loc=mean, scale=std_dev)

# Valores para o eixo x
x_values = np.linspace(0, 16, 1000)

# Calcula os valores da densidade de probabilidade e da função de sobrevivência
pdf_values = norm.pdf(x_values, loc=mean, scale=std_dev)
sf_values = norm.sf(x_values, loc=mean, scale=std_dev)

# Defina o intervalo desejado (por exemplo, entre x=6 e x=10)
interval_mask = (x_values >= 2) & (x_values <= 6)

# Gera o gráfico
plt.plot(x_values, pdf_values, label='Distribuição normal')
plt.fill_between(x_values[interval_mask], 0, pdf_values[interval_mask], color='skyblue', alpha=0.3)
plt.axvline(6, color='red', linestyle='-', label='x = 6')
plt.legend(loc='upper right', fontsize='small')
plt.title('Distribuição normal')
plt.show()
```

#### Qual a probabilidade de tirar um objeto que o peso é maior que 6 quilos?

```python
norm.sf(6, 8, 2)
# 1 - norm.cdf(6, 8, 2) # mesmo resultado da função acima
```

```python
# Parâmetros da distribuição normal
mean = 8
std_dev = 2

# Ponto específico
x_point = 6
sf_value_at_x_point = norm.sf(x_point, loc=mean, scale=std_dev)

# Valores para o eixo x
x_values = np.linspace(0, 16, 1000)

# Calcula os valores da densidade de probabilidade e da função de sobrevivência
pdf_values = norm.pdf(x_values, loc=mean, scale=std_dev)
sf_values = norm.sf(x_values, loc=mean, scale=std_dev)

# Defina o intervalo desejado (por exemplo, entre x=6 e x=14)
interval_mask = (x_values >= 6) & (x_values <= 14)

# Gera o gráfico
plt.plot(x_values, pdf_values, label='Distribuição normal')

# Preenche a área sob a curva da função de sobrevivência apenas no intervalo desejado e abaixo da linha da distribuição normal
plt.fill_between(x_values[interval_mask], 0, pdf_values[interval_mask], color='skyblue', alpha=0.3) # , label='Area under SF (6 > x < 12)'

# Adiciona uma linha vertical para x=6
plt.axvline(6, color='red', linestyle='-', label='x = 6')

# Move a legenda para um pouco acima do centro da parte direita
#plt.legend(loc='center right', bbox_to_anchor=(1, 0.5), fontsize='small')

# Move a legenda para partesuperior direita
plt.legend(loc='upper right', fontsize='small')

plt.title('Distribuição normal')
#plt.xlabel('X')
plt.grid(True)
plt.show()
```

#### Qual a probabilidade de tirar um objeto que o peso é menor que 6 ou maior que 10 quilos?

```python
norm.cdf(6, 8, 2) + norm.sf(10, 8, 2)
```

```python
# Parâmetros da distribuição normal
mean = 8
std_dev = 2

# Ponto específico
x_point = [6, 10]
sf_value_at_x_point = norm.sf(x_point, loc=mean, scale=std_dev)

# Valores para o eixo x
x_values = np.linspace(0, 16, 1000)

# Calcula os valores da densidade de probabilidade e da função de sobrevivência
pdf_values = norm.pdf(x_values, loc=mean, scale=std_dev)
sf_values = norm.sf(x_values, loc=mean, scale=std_dev)

# Defina o intervalo desejado
interval_mask1 = (x_values >= 2) & (x_values <= 6)
interval_mask2 = (x_values >= 10) & (x_values <= 14)

# Gera o gráfico
plt.plot(x_values, pdf_values, label='Distribuição normal')
plt.fill_between(x_values[interval_mask1], 0, pdf_values[interval_mask1], color='skyblue', alpha=0.3)
plt.fill_between(x_values[interval_mask2], 0, pdf_values[interval_mask2], color='skyblue', alpha=0.3)
plt.axvline(6, color='red', linestyle='-', label='x = 6')
plt.axvline(10, color='orange', linestyle='-', label='x = 10')
plt.legend(loc='upper right', fontsize='small')
plt.title('Distribuição normal')
plt.show()
```
