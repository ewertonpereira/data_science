### Distribuição T student

#### Importando bibliotecas

```python
from scipy.stats import t, norm
import matplotlib.pyplot as plt
import numpy as np
```

#### Média do salário dos cientistas de dados = R$75,00 por hora  amostra com 9 funcionários e desvio padrão = 10

* Qual a probabilidade de selecinar um cientista de dados e o salário ser menor que R$80,00 por hora?

```python
t.cdf(1.5, 8)
```

```python
# Parâmetros da distribuição normal
mean = 75
std_dev = 10

# Ponto específico
x_point = 80

# Valores para o eixo x
x_values = np.linspace(0, 150, 1000)

# Calcula os valores da densidade de probabilidade e da função de sobrevivência
pdf_values = norm.pdf(x_values, loc=mean, scale=std_dev)

# Defina o intervalo desejado (por exemplo, entre x=6 e x=10)
interval_mask = (x_values >= 40) & (x_values <= 80)

# Gera o gráfico
plt.plot(x_values, pdf_values, label='Distribuição normal')
plt.fill_between(x_values[interval_mask], 0, pdf_values[interval_mask], color='skyblue', alpha=0.3)
plt.axvline(80, color='red', linestyle='-', label='x = 80')
plt.legend(loc='upper right', fontsize='small')
plt.title('Distribuição normal')
plt.show()
```
